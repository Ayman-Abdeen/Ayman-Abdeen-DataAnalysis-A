```python
df_Unplanned = Unplanned[['siteid','category', 'subcategory','area','duration', 'dateid', 'controller']]
df_Unplanned
```

```python
df_Planned = Planned[['Site','Category', 'SubCategory','Region','Duration', 'DateOfOutage' ,'Controller']]
df_Planned
```

```python
columnsName = ['Site','Category', 'SubCategory','Region','Duration', 'DateOfOutage' ,'Controller']

df_Planned.columns = columnsName
df_Unplanned.columns = columnsName
```

```python
frames =[df_Unplanned,df_Planned]
allrecords = pd.concat(frames)
allrecords
```

```python
len( allrecords[allrecords['Controller'].str.contains("LTE")] )
```

```python
new_df = allrecords[~allrecords["Controller"].str.contains("LTE")]
new_df
```
