```python
df_Unplanned = Unplanned[['siteid','category', 'subcategory','area','duration', 'dateid', 'controller']]
df_Unplanned
```

```python
df_Planned = Planned[['Site','Category', 'SubCategory','Region','Duration', 'DateOfOutage' ,'Controller']]
df_Planned
```

```python
columnsName = ['Site','Category', 'SubCategory','Region','Duration', 'DateOfOutage' ,'Controller']

df_Planned.columns = columnsName
df_Unplanned.columns = columnsName
```

```python
frames =[df_Unplanned,df_Planned]
allrecords = pd.concat(frames)
allrecords
```

```python
len( allrecords[allrecords['Controller'].str.contains("LTE")] )
```

```python
new_df = allrecords[~allrecords["Controller"].str.contains("LTE")]
new_df
```

```python
new_df['Duration'].div(60)
```

```python
new_df['Duration_Hr'] = new_df['Duration'].div(60)
```

```python
new_df
```

```python
print(xlrd.xldate_as_datetime(44793, 0).date())
```

```python
new_df['DateOfOutage'] = new_df['DateOfOutage'].apply(lambda x : xlrd.xldate_as_datetime(x, 0).date())
```

```python
new_df
```

```python
 new_df['SubCategory'].str.contains('ROT')
```

```python
new_df['ParkedOrROT']='Other reason'
```

```python
new_df.loc[new_df['Duration'] == 1440 , 'ParkedOrROT'] = "Parked"
```

```python
new_df.loc[new_df['SubCategory'] == 'Under_ROT', 'ParkedOrROT'] = 'ROT'
```

```python
new_df['NCategory']=new_df['Category']
```

```python
new_df.loc[new_df['NCategory'].str.contains('Transmission'), 'NCategory'] =  "Transmission"
```

```python
new_df['NCategory'] = new_df['NCategory'].apply(lambda x: "Non-technical" if "Damaged_Stolen" in x or "Missing_Spare_parts"  in x or "Dependency_Huawei"  in x or "HUAWEI_FM"  in x or "Under_ROT"  in x else x)
```

```python
new_df
```

```python
groupeddf = new_df.groupby("NCategory").count()
groupeddf
```

```python
SumOfDuration_HrNCategory=new_df.groupby("NCategory").sum("Duration_Hr")
SumOfDuration_HrNCategory
```

```python
SumOfDuration_HrNCategory["Duration_Hr"].sum()
```

```python
SumOfDuration_HrNCategory["Total_Lost_Hours"]=np.abs(SumOfDuration_HrNCategory["Duration_Hr"].sub(SumOfDuration_HrNCategory["Duration_Hr"].sum()))
```

```python
SumOfDuration_HrNCategory
```

```python
SumOfDuration_HrParkedOrROT=new_df.groupby("ParkedOrROT").sum("Duration_Hr")
SumOfDuration_HrParkedOrROT
```

```python
SumOfDuration_HrParkedOrROTCount=new_df.groupby("ParkedOrROT").count()
SumOfDuration_HrParkedOrROTCount
```

```python
SumOfDuration_HrRegion=new_df.groupby("Region").sum("Duration_Hr")
SumOfDuration_HrRegion
```

```python
SumOfDuration_HrDateOfOutage=new_df.groupby("DateOfOutage").sum("Duration_Hr")
SumOfDuration_HrDateOfOutage
```
